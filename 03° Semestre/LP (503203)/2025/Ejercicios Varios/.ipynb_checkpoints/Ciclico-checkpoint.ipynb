```python
def esta(c,s):
    for i in range(len(s)):
        if(c==s[i]):
            return True

    return False


def ciclico(n): 
    s=str(n)
    digitos=len(s)
    for i in range(1,digitos+1):
        k=str(n*i)
        for j in range(len(k)):
            if(not esta(k[j],s)):
                return False
    return True

def main():
    n=int(input())
    if(ciclico(n)):
        print(n,"es ciclico")
    else:
        print(n,"no es ciclico")

main()
```
