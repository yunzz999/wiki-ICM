```python
'''
Observaciones: Sección exclusiva para el comentarios de él/la revisor/a.

Puntaje : Sección exclusiva para el comentarios de él/la revisor/a.
'''

#  CÓDIGO DEL PROBLEMA 1
def leevalida(vmin,vmax,txt):
    n=int(input(txt))
    while(not vmin<=n<=vmax):
        n=int(input(f"Error el valor ingresado debe estar entre {vmin} y {vmax}.\n {txt}"))
    return n

    
def fibo(n):
    if(n==1):
        return 0
    if(n==2):
        return 1
    f1=0
    f2=1
    temp=0
    for i in range(2,n+1):
        temp=f2
        f2=f2+f1
        f1=temp
    return f2

def main():
    n=leevalida(2,float('inf'),"n: ")
    aureo=(1+5**0.5)/2
    x=fibo(n)/fibo(n-1)
    return abs(x-aureo)

main()
```

```python
'''
Observaciones: Sección exclusiva para el comentarios de él/la revisor/a.

Puntaje : Sección exclusiva para el comentarios de él/la revisor/a.
'''
#  CÓDIGO DEL PROBLEMA 2
def leevalida(vmin,vmax,txt):
    n=int(input(txt))
    while(not vmin<=n<=vmax):
        n=int(input(f"Error el valor ingresado debe estar entre {vmin} y {vmax}.\n {txt}"))
    return n

def digitos(n):
    sum=0
    while(n>=1):
        sum+=1
        n=n/10
        
    return sum

def invertido(m):
    k=digitos(m)
    sum=0
    while(m>=1):
        sum+=(m%10)*10**(k-1)
        k=k-1
        m=m//10
    return sum
def main():
    m=leevalida(100,10**12,"M: ")
    inv=invertido(m)
    return inv

main()
```

```python
'''
Observaciones: Sección exclusiva para el comentarios de él/la revisor/a.

Puntaje : Sección exclusiva para el comentarios de él/la revisor/a.
'''

#  CÓDIGO DEL PROBLEMA 3

def leevalida(vmin,vmax,txt):
    n=int(input(txt))
    while(not vmin<=n<=vmax):
        n=int(input(f"Error el valor ingresado debe estar entre {vmin} y {vmax}.\n {txt}"))
    return n


def valorterna(simb,a,b):
    if(simb==1):
        return a+b
    elif(simb==2):
        return a-b
    elif(simb==3):
        return a*b
    elif(simb==4 and b!=0): #que pasa si b=0
        return a/b

def main():
    n=leevalida(1,float('inf'),"t: ")
    for i in range(n):
        a=float(input(f"a{i+1}: "))
        b=float(input(f"b{i+1}: "))
        o=leevalida(1,4,f"o{i+1}: ")
        maximo=-1
        x=valorterna(o,a,b)
        if(x>maximo):
            maximo=x
    return x
main()
```
