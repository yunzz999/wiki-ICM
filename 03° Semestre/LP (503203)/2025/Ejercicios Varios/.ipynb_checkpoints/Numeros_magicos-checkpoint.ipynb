```python
def is_magic(n):
    x=str(n**3)
    sum=0
    for i in range(len(x)):
        sum+=int(x[i])
    if(sum==n):
        return True
    else:
        return False

def main():
    n=int(input("n: "))
    #validar
    s=""
    for i in range(1,n+1):
        if(is_magic(i)):
            s+=str(i)+", "

    print(s)
main()
```
