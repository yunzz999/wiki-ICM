```python
def leevalida(n,vmin,vmax,txt):
    while(n>=vmax or n<vmin):
        n=int(input(txt))
    return n

def is_magic(n):
    x=str(n**3)
    sum=0
    for i in range(len(x)):
        sum+=int(x[i])
    if(sum==n):
        return True
    else:
        return False

def main():
    n=int(input("n: "))
    n=leevalida(n,1,float('inf'),"Error, nÃºmero fuera de rango.")
    #validar
    for i in range(1,n+1):
        if(is_magic(i)):
            print(i)

main()
```
