```python
def leevalida(n,vmin,vmax,txt):
    while(n<vmin or n>vmax):
        n=int(input(txt))
    return n

def ultimodig(n):
    return n%10

def sol(n):
    if(ultimodig(n)!=0):
        n=n-1
    else:
        n=n//10
    return n

def main():
    n=int(input("n: "))
    n=leevalida(n,2,1009,"Error ingrese denuevo: ")
    k=int(input("k: "))
    k=leevalida(k,1,50,"Error ingrese denuevo: ")
    for i in range(k):
        n=sol(n)
    print(n)

main()
```
