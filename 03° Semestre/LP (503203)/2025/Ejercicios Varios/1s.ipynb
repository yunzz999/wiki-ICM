```python
def sol(n):
    s="1"
    while(int(s)%n!=0):
        s+="1"
    return int(s)

def main():
    t=int(input("t: "))
    for i in range(t):
        n=int(input("n: "))
        if(n%2==0):
            print(n,"es par")
        elif(n%5==0):
            print(n,"es div por 5")
        else:
            print(sol(n)//n,"-",sol(n))

main()
```
