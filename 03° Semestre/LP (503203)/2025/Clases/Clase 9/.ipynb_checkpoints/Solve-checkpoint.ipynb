```python
import numpy as np 

def leevalida(m,M,txt):
    x=int(input(txt))
    while(not m<= x <= M):
        x=int(input("Error, ingrese denuevo: "))
    return x

m=leevalida(1,20,"m: ")
l=[]
mini=float("inf")
posmini=-1
maxi=-float("inf")
posmaxi=-1
for i in range(m):
    x=int(input())
    l.append(x)
    if(x>maxi):
        maxi=x
        posmaxi=i
    if(x<mini):
        mini=x
        posmini=i

l=sorted(l)
print(f"menor ={mini} en posicion={posmini+1}")
print(f"mayor ={maxi} en posicion={posmaxi+1}")
print()



```
