```python
'''
Observaciones: Sección exclusiva para el comentarios de él/la revisor/a.

Puntaje : Sección exclusiva para el comentarios de él/la revisor/a.
'''

#  CÓDIGO DEL PROBLEMA 3

def leevalida(vmin,vmax,txt):
    n=int(input(txt))
    while(not vmin <= n <= vmax):
        n=int(input(f"Error el valor debe estar entre {vmin} y {vmax}\n{txt}"))
    return n

def absol(x): #Funcion valor absoluto
    if(x>=0):
        return x
    else:
        return -x
        
def calculo_F(n):  
    #Suma de las fuerzas
    sumf=0
    
    #Primeros datos
    d=float(input(f"Distancia {1}: "))
    h=float(input(f"Altura {1}: "))
    
    #Altura anterior
    antmur=h

    #Datos
    for i in range(1,n):
        d=float(input(f"Distancia {i+1}: "))
        h=float(input(f"Altura {i+1}: "))
        
        #Calculo
        sumf+=d+absol(h-antmur) #diferencias de altura debe ser positivo
        antmur=h                #se reemplaza la altura anterior por la actual

    return sumf


n=leevalida(1,100,"N: ")  
print(calculo_F(n))

```
