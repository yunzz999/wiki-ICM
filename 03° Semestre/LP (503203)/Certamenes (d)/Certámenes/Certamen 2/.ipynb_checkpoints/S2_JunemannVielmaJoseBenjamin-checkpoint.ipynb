```python
'''
Observaciones: Sección exclusiva para el comentarios de él/la revisor/a.

Puntaje : Sección exclusiva para el comentarios de él/la revisor/a.
'''
#  CÓDIGO DEL PROBLEMA 1

def leevalida1(m,M,txt):
    x=int(input(txt))
    while(not m<= x <= M):
        x=int(input("Error, ingrese denuevo: "))
    return x

def leevalida2(m,M,txt):
    x=float(input(txt))
    while(not m<= x <= M):
        x=float(input("Error, ingrese denuevo: "))
    return x    

nombres=[]
prueba1=[]
prueba2=[]
prueba3=[]
promedios=[]

n=leevalida1(1,49,"n: ")
for i in range(n):
    nombre=input(f"Nombre del alumno {i+1}: ")
    nombres.append(nombre)
    p1=leevalida2(1,7,f"Nota prueba 1:")
    p2=leevalida2(1,7,f"Nota prueba 2:")
    p3=leevalida2(1,7,f"Nota prueba 3:")
    promedios.append((p1+p2+p3)/3)
    prueba1.append(p1)
    prueba2.append(p2)
    prueba3.append(p3)

print()
for i in range(n):
    print(f"{nombres[i]} : {promedios[i]}")
    
print()
print(f"Promedio prueba 1: {(sum(prueba1))/n}")
print(f"Promedio prueba 2: {sum(prueba2)/n}")
print(f"Promedio prueba 3: {sum(prueba3)/n}")
print(f"Promedio General: {(sum(prueba1)+sum(prueba2)+sum(prueba3))/(3*n)}")

    
```

```python
'''
Observaciones: Sección exclusiva para el comentarios de él/la revisor/a.

Puntaje : Sección exclusiva para el comentarios de él/la revisor/a.
'''
#  CÓDIGO DEL PROBLEMA 2
import numpy as np 

def leevalida(m,M,txt):
    x=int(input(txt))
    while(not m<= x <= M):
        x=int(input("Error, ingrese denuevo: "))
    return x

def creamat(n,m,nM):
    R=np.zeros((n,m),dtype=int)
    for i in range(n):
        for j in range(m):
            R[i,j]=leevalida(0,100,f"{nM}[{i+1},{j+1}]=")
    return R


def solve2(A,n,m):
    g=[]  #elementos no repetidos 
    l=[]  # todos los elementos
    c=[]  #cantidad de elementos
    for i in range(n):
        for j in range(m):
            l.append(A[i,j]) #añado tdos los elementos
            if(A[i,j] not in g):
                g.append(A[i,j]) #añado solo si no habia uno antes
    
    g=sorted(g)   #ordeno  #O (NM log NM) 
    for y in g: 
        c.append(l.count(y)) #cuento elementos y agrego
    
    R=np.array([g,c]) #matriz resultante
    return R 

#O(NM log NM)
    
def solve(A,n,m):
    l=[]
    c=[]
    for i in range(n):
        for j in range(m):
            if(A[i,j] not in l):
                l.append(A[i,j])  #añado si no esta 
                c.append(1)       # sumo 1 a su contador
            else:
                c[l.index(A[i,j])]+=1    #encuentro su indice donde le corresponde en los contadores y sumo 1

    R=np.array([l,c])
    return R

# O(NM*(NM)) horrible xd


n=leevalida(2,999,"n: ")
m=leevalida(2,999,"m: ")
DU=creamat(n,m,"DU")
R=solve(DU,n,m)
K=solve2(DU,n,m)
print(R)
print(K)
```
