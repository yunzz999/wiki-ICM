```python
#Problema 1

def leevalida(m,M,txt):
    s=(input(txt))
    x=len(s)
    while(not m<= x <= M):
        s=(input("Error, ingrese denuevo: "))
        x=len(s)
    return s


def eliminado(c):
    s="aeiou"
    if c in s:
        return "-"
    if c==" ":
        return ""

    else:
        return c


s=leevalida(1,1000,"Ingrese su cadena: ")
s=s.lower()
sol=""
for k in s:
    sol+=eliminado(k)

print(sol)



```

```python
#Problema 2

def es_lider(k,l):
    if(k==len(l)-1):
        return False
        
    for i in range(k+1,len(l)):
        if(l[i]>l[k]):
            return False

    return True


def es_primo(p):
    if(p==0):
        return False
    if(p==1):
        return True
    for i in range(2,p):
        if(p%i==0):
            return False
    return True
    
def primos_n(n):
    l=[]
    for i in range(n+1):
        if(es_primo(i)):
            l.append(i)
    return l


n=int(input("n: "))
numeros=[]
maxi=-1
for i in range(n):
    x=int(input(f"n{i+1}=")) 
    if(x>maxi):
        maxi=x
    numeros.append(x)

sol=[]
primos=primos_n(maxi)
for i in range(n):
    s=[]
    for p in primos:
        if(p not in s and numeros[i]-p in primos):
            print(f"Pareja de primos de {numeros[i]} son: {p} {numeros[i]-p}")
            sol.append(p)
            s.append(p)
            s.append(numeros[i]-p)
            sol.append(numeros[i]-p)
print()
for i in range(len(sol)):
    if(es_lider(i,sol)):
        print(sol[i],end=" ")
print()
print(sol)
```

```python
#Problema 3

import numpy as np

def leevalida(m,M,txt):
    x=int(input(txt))
    while(not m<= x <= M):
        x=int(input("Error, ingrese denuevo: "))
    return x



n=leevalida(2,10,"n: ")
A=np.zeros((n,n))
for i in range(n):
    A[0][i]=leevalida(-9,9,f"A[0][{i+1}]=")

for i in range(1,n):
    for j in range(n):
        if(j>=i):
            t=A[i-1][j-1]+A[i-1][j]
            if(t>=10):
                t=t%10
            A[i][j]=t

print(A)
                




```
